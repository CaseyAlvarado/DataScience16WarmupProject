# Look at the survival rate based on different features (e.g. males versus females, young versus old, etc.)

```python
females = raw_data.Survived[raw_data.Sex == 'female']
males = raw_data.Survived[raw_data.Sex == 'male']
```

```python
females_hist = thinkstats2.Hist(females)
males_hist = thinkstats2.Hist(males)

width = 0.45

thinkplot.PrePlot(2)
thinkplot.Hist(females_hist, align = 'left', width = width, label = 'females')
thinkplot.Hist(males_hist, align = 'right', width = width, label = 'males')
thinkplot.Show(xlabel = 'female survived or died', ylabel = 'number of people')
```

The graph above has two extemes, zero or one, where zero means the people died or where one means the people survived. This data shows that more female survived the titanic and more males died. This makes sense since they were putting women and children on the boat before men. 

```python
children = raw_data.Survived[raw_data.Age <= 13]
adults = raw_data.Survived[(raw_data.Age > 13) | (raw_data.Age <= 50)]
elderly = raw_data.Survived[raw_data.Age > 50]


children_hist = thinkstats2.Hist(children)
adults_hist = thinkstats2.Hist(adults)
elderly_hist = thinkstats2.Hist(elderly)

width = 0.1
thinkplot.PrePlot(3)
thinkplot.Hist(children_hist, align = 'left', width = width, label = 'children under the age of 13')
thinkplot.Hist(males_hist, align = 'center', width = width, label = 'adults between the ages of 13 - 50')
thinkplot.Hist(elderly_hist, align = 'right', width = width, label = 'elderly over the ages of 50')
thinkplot.Show(xlabel = 'survival or death', ylabel = 'number of people')
```

```python
paid_alot = raw_data.Survived[raw_data.Fare >= 50]
paid_less = raw_data.Survived[raw_data.Fare < 50]

paid_alot_hist = thinkstats2.Hist(paid_alot)
paid_less_hist = thinkstats2.Hist(paid_less)

width = 0.2 
thinkplot.PrePlot(2)
thinkplot.Hist(paid_alot_hist, align = 'left', width = width, label = 'people that paid more than 80 dollars for a ticket')
thinkplot.Hist(paid_less_hist, align = 'right', width = width, label = 'people that paid less than 80 dollars for a ticket')
thinkplot.Show(xlabel = 'died = 0, survived = 1', ylabel = 'number of people')
```

This plot shows that the majority of the people that survived were children.

```python
thinkplot.Scatter(raw_data.Survived, raw_data.Fare)
thinkplot.Show(xlabel = 'survived, 0 = death, 1 = survived', ylabel = 'number of people')
```

```python
    df = raw_data
    df = df.dropna(subset = ['Survived', 'Fare'])
    bins = numpy.arange(5, 50, 5)
    indices = numpy.digitize(df.Fare, bins)
    groups = df.groupby(indices)
    
    for i, group in groups: 
        print (i, len(group))
    
    fares = [group.Fare.mean() for i, group in groups]
    cdfs = [thinkstats2.Cdf(group.Survived) for i, group in groups]

    for percent in [75, 50, 25]: 
        survived = [cdf.Percentile(percent) for cdf in cdfs]
        label = '%dth' % percent
        
        thinkplot.Plot(  fares, survived); 
thinkplot.Show(xlabel = 'survived', ylabel = 'fares')
```

According to the binning plot above, the graph is flat. Let's take the Spearman and Pearson's correlation to see if these variables correlate in any way.

```python
thinkstats2.Corr(raw_data.Survived, raw_data.Fare)
```

```python
thinkstats2.SpearmanCorr(raw_data.Survived, raw_data.Fare)
```

There is a small difference in the result from these two correlations. The Spearman correlation is higher, this could be because the rate relating survivor and fares could be non linear. Or because the Spearman correlation works to rank the people, the outliers might be skewing the fare data. The majority of the people might have paid a small fare, but wealthier people might be skewing the data. To investigate this further, we should plot the histogram of the fares.

```python
# convert from float to int
raw_data.Fare = raw_data.Fare.astype(int)

raw_data.Fare.hist(bins=70)
      

fare_hist = thinkstats2.Hist(raw_data.Fare)
thinkplot.Hist(fare_hist, label = 'fare histogram')
thinkplot.Show(xlabel = 'fare values', ylabel = 'number of people that paid that fare') 
```

As you can see in the plot above, the majority of the people paid between 0 and 100 dollars. However, there were people that paid around 500 follars. This skews the results so that the Pearson correlation is not accurate due to these outliers. There does seem to be a small positive correlation between fare values and survivors.
